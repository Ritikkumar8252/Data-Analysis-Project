```python
# IMPORT  ALLTHE NECESSARY LIBRARY

import pandas as pd  #here np is alias (shart form for  using)
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#  import  csv file
# read file using "pandas"

df = pd.read_csv('/content/data-export (1).csv')
```

```python
df.head() #first  5 row data
#  here we see the data is not cleaned
# column name is not correct all are  unntitled
```

```python
df.columns =df.iloc[0] # it means row of zero index  become the nname or, title of column
df.head()
```

```python
#  above title ofcolumn is correct but row no. "0"  is  also same as title so drop  that row
df = df.drop(index=0).reset_index(drop=True)
# df = df.drop(index=0): This line removes the row with index 0 from the DataFrame.
# .reset_index(drop=True): After dropping a row, the index of the DataFrame is no longer sequential.
#         This part of the code resets the index so that it starts from 0 again.
#         drop=True prevents the old index from being added as a new column.
df.head()
```

```python
# change the name of column title according to your own
df.columns = ["Session primary channel group","Date_hour",	"Users","Sessions",	"Engaged sessions",	"Average engagement time per session"	,"Engaged sessions per user",	"Events per session",	"Engagement rate"	,"Event count"]
# NOTE:- "JITNA COLUMN HAI UTNA  COLUMN KA NAME LIKNA OTHER WISE VALUE ERROR"
df.head()
```

```python
# check every colummn Data of which datatype
df.info()
```

```python
# ALL COLUMN DATA ARE OBJECT TYPE SO  WE NEED TO CHANGE THEM.....
# COLUMN 1 :- are sting tye that  is object  type
# COLUMN 2 :- need to change in date time format
# ALL_OTHERS_COLUMNS  :-change in numerical data format
df["Date_hour"] = pd.to_datetime(df["Date_hour"], format='%Y%m%d%H', errors='coerce')
# pd.to_datetime():- converts strings (or numbers) into datetime objects
df.head()
```

```python

# Convert all columns except 'Date_hour' to numeric
for col in df.columns:
    if col not in ['Date_hour', 'Session primary channel group']:
        df[col] = pd.to_numeric(df[col], errors='coerce')
df.head()

```

```python
df.info()
```

```python
# date and time in to two different column

df['Hour'] = df['Date_hour'].dt.hour
# .dt
#   The .dt accessor allows you to extract specific datetime parts (like hour, day, month, year) from a datetime64[ns] column in pandas.

# .hour
#   This pulls out just the hour part (023) from each datetime.

# df['Hour'] = ...
#   The extracted hour values are stored in a new column named "Hour"
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#  ------------------------------->NOW  DATA IS CLEANED<--------------------------------
# Q 1. WHAT PATTER OR TREND THAT CAN YOU OBSERVE IN WEBSITE SESSION AND USER OVER TIME ?
#     ANS-: Its asking you to look at how your website traffic changes
#            over time and describe any trends or patterns.
#             Sessions:- Number of visits to your website (can include repeat visitors).
#              Users:- Number of unique people visiting your website.
#               Time component:- Could be hours, days, weeks, or months, depending on your dataset (Date_hour, Date, etc.).
# Why:- Line charts are ideal for time series because they show growth, decline, or stability over time


# Optional: Use seaborn style
sns.set_style('whitegrid') #'whitegrid' adds a light grid background, which makes it easier to read trends.

plt.figure(figsize=(12,6)) #figsize=(12,6) sets the width and height of the plot in inches.

# Plot Sessions
sns.lineplot(data=df, x='Date_hour', y='Sessions', label='Sessions')
      # data=df tells seaborn which DataFrame to use.
       # x='Date_hour'  X-axis shows the datetime column.
        # y='Sessions'  Y-axis shows the number of sessions.
         # label='Sessions'  Name of the line, which will appear in the legend.

# Plot Users
sns.lineplot(data=df, x='Date_hour', y='Users', label='Users') #This allows you to compare sessions vs users on the same chart.

# Titles and labels
plt.title('Website Sessions and Users Over Time')
plt.xlabel('Date and Hour')
plt.ylabel('Count')
plt.legend() #Shows the legend on the chart, which explains which line is sessions and which is users.
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Optional: Aggregate daily
# ---->>  df_daily = df.groupby(df['Date_hour'].dt.date).sum().reset_index()
# df['Date_hour'].dt.date works fine for grouping (it extracts just the date).
# But sum() tries to sum all columns, including your datetime column, which is not numeric.

# Aggregate numeric columns by day
df_daily = df.groupby(df['Date_hour'].dt.date)[['Sessions', 'Users']].sum().reset_index()

# Set seaborn style
sns.set_style('whitegrid')

# Create figure
plt.figure(figsize=(12,6))

# Plot Sessions
sns.lineplot(data=df_daily, x='Date_hour', y='Sessions', label='Sessions')

# Plot Users
sns.lineplot(data=df_daily, x='Date_hour', y='Users', label='Users')

# Titles and labels
plt.title('Website Sessions and Users Over Time')
plt.xlabel('Date')
plt.ylabel('Count')
plt.legend()  # Shows which line is Sessions and which is Users
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# TOTAL USER  BY CHANNEL
plt.figure(figsize=(8,5))

# Bar plot
sns.barplot(data=df, y='Users', x='Session primary channel group', palette='viridis')

# Titles and labels
plt.title('Total Users by Channel')
plt.xlabel('Total Users')
plt.ylabel('Channel')
plt.tight_layout()
plt.show()
```

```python

# AVERAGE ENGAGEMENT TIME
# Average Engagement Time = Total time spent by all users  Number of sessions.
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))

# Vertical bar plot
sns.barplot(data=df, x='Session primary channel group', y='Average engagement time per session', palette='viridis')

# Titles and labels
plt.title('Average Engagement Time by Channel')
plt.xlabel('Channel')
plt.ylabel('Average Engagement Time (minutes)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ENGAGEMENT RATE DISTRIBUTION BY CHANNEL
# A box plot (or whisker plot) is a statistical visualization that shows the distribution of a numeric variable  in your case,
#  Engagement rate  for each category (channel).
plt.figure(figsize=(10,6))

# Box plot
sns.boxplot(data=df, x='Session primary channel group', y='Engagement rate', palette='viridis')
# "What a box plot shows
  # For each channel:
  # Box (interquartile range, IQR):
  # Shows the middle 50% of the data (25th to 75th percentile).
    # Median line:
    # The line inside the box shows the median engagement rate.
    # Whiskers:
    # Extend to the lowest and highest values within 1.5IQR from the quartiles.
    # Outliers:
    # Points outside the whiskers are plotted individually. These are unusually high or low engagement rates."
```
